```
import numpy as np
import pandas as pd
import json
import requests
```

```
def get_businesses(location, term, api_key):
    headers = {'Authorization': 'Bearer %s' % api_key}
    url = 'https://api.yelp.com/v3/businesses/search'

    data = []
    for offset in range(0, 1000, 50):
        params = {
            'limit': 50, 
            'location': location.replace(' ', '+'),
            'term': term.replace(' ', '+'),
            'offset': offset
        }

        response = requests.get(url, headers=headers, params=params)
        if response.status_code == 200:
            data += response.json()['businesses']
        elif response.status_code == 400:
            print('400 Bad Request')
            break

    return data
```

```
api_key = 'Yelp Api Key'
headers = {'Authorization': 'Bearer %s' % api_key}
term='Doctor'
location='Washington, DC'
all_data=get_businesses(location, term, api_key)
```

```
uid=[]
disp_address=[]
name=[]
review_count=[]
url=[]
overall_rating=[]
for businesses in all_data:
  name.append(businesses['name'])
  uid.append(businesses['id'])
  review_count.append(businesses['review_count'])
  url.append(businesses['url'])
  disp_address.append(businesses['location']['display_address'])
  overall_rating.append(businesses['rating'])
    
dict = {'Name': name, 'UID': uid, 'Address': disp_address, 'Reviews_count':review_count,'URL':url,'Overall_Rating':overall_rating}      
data_frame = pd.DataFrame(dict)   
```

```
data_frame['URL']=data_frame['URL'].str.split('?').apply(lambda x: x[0])
```

```
data_frame
```

```
def doc_function(get_id):
  rating=[]
  review=[]
  ids=[]
  Date=[]
  New_Url="https://api.yelp.com/v3/businesses/"+get_id+"/reviews"
  params = {
            'location': location.replace(' ', '+'),
            'term': term.replace(' ', '+'),
        }
  data = requests.get(New_Url, headers=headers, params=params)
  per_person_data=json.loads(data.text)
  for data in per_person_data['reviews']:
    rating.append(data['rating'])
    review.append(data['text'])
    ids.append(get_id)
    Date.append(data['time_created'])
  dict = {'Review': review, 'UID': ids, 'Rating': rating, 'Date':Date }          
  return dict  
  
```

```
final_df = pd.DataFrame(columns=['Review','UID','Rating','Date'])#empty data frame
for set_id in data_frame['UID']:
  final_df=final_df.append(pd.DataFrame(doc_function(set_id)),ignore_index=True)
```
